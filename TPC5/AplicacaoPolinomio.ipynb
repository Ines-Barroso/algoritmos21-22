* Resumo: Este algoritmo consiste num conjunto de funções que manipula um polinómio. O menu dá oportunidade ao utilizador de escolher o que quer fazer, aparecendo no início e após corrida uma função. Para efetuar uma das opções foi necessário importar o modulo _matplotlib.pyplot_ . O polinómio é manipulado como uma lista de tuplos, sendo cada tuplo um termo do polinómio.
    

```python
def menu():
    print(""" Aplicação Polinómio
(1)Criar polinómio
(2)Calcular polinómio criado
(3)Calcular o polinómio para diferentes valores de x 
(4)Simplificar o polinómio 
(5)Derivar o polinómio
(6)Gráfico do polinómio
(0)Sair""")

import matplotlib.pyplot as plt

def verTermo(termo):
    if termo[1] == 0:
        return str(termo[0])
    else:
        return (str(termo[0]) + "x^" + str(termo[1]))
    
def verPolinomio(p):
    res = ""
    for t in p:
        if res == "" :
            res = verTermo(t)
        else:
            res = res + " + " + verTermo(t)
    return res

def chaveOrd(t):
    _,e = t 
    return e

def ordenarPol(p):
    p.sort(reverse = True, key=chaveOrd)

def criarPoli():
    poli = []
    n_termos = int(input("Um polinómio é composto por termos (exemplo de termo: 4x^3). \n Quantos termos tem o polinómio que quer criar? "))
    for i in range (n_termos):
        coef = int(input("Qual o coeficiente? "))
        exp = int(input("Qual o expoente? "))
        termo = (coef,exp)
        poli.append(termo)
    return poli 

def calPol(p,x):
    resultado = 0
    for t in p:
        c, e = t
        resultado = resultado + (c*(x**e))
    return  resultado

def tabela(p,n):
    for i in range(n+1):
        print (str(i) + " :: " + str (calPol(p,i)))
        
def simplificarPol(p):
    ordenarPol(p)
    resLista = []
    for i in range(1, len(p)):
        c, e = p[i]
        c0, e0 = p[i-1]
        if e0 == e:
            resLista.append((c+c0, e))
        else: 
            if e0 > e:
                if i == 1:
                    resLista.append((c0, e0))
                else:
                    resLista.append((c,e))
    return resLista        

def derivadaPol(p):
    res =[]
    for (c, e) in p:
        if e > 1 :
            res.append(((c*e), e-1))
    return res  

def graficoPol(p):
    x = [i for i in range(-100,100)]
    y = []
    for n in x:
        y.append(calPol(p, n))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.title('Polinomio')

```

```python
op = 1    
while op != "0":
    menu()
    op = input ("Qual é a sua opção? ")
    if op == "1":
        poli = criarPoli()
        print ("O polinómio criado foi: ", verPolinomio(poli))
    elif op == "2":
        x = int(input("Qual o valor de x?"))
        cp = calPol(poli, x)
        print ("A soma do polinómio para x=", x, " é:", str(cp))
    elif op == "3":
        n = int(input("Até que valor de x quer calcular o polinómio.\n Por exemplo, se quiser até 20, coloque 20 e os valores de x irão variar de 0 a 20.\n"))
        xcp = tabela(poli, n)
        print (xcp)      
    elif op == "4":
        sp = simplificarPol(poli)
        print ("Polinómio simplificado é: ", verPolinomio(poli))       
    elif op == "5":
        dp = derivadaPol(sp)
        print ("Derivada de ",verPolinomio(poli),"é: ", verPolinomio(dp))
    elif op == "6":        
        graficoPol(sp)
    elif op == "0":
        print("Até à próxima...")
    else :
        print("Opção inválida.")
      
```
