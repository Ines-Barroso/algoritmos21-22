* Resumo: Este algoritmo consiste num conjunto de funções que manipulam frações. O menu dá oportunidade ao utilizador de escolher o que quer fazer, aparecendo no início e após corrida uma função.
    

```python
def menu():
    print(""" Aplicação para frações
(1)Criar fração
(2)Simplificar a fração que criou
(3)Somar duas frações
(4)Subtrair duas frações
(5)Multiplicar duas frações
(6)Dividir duas frações
(7)Gerar uma lista de n frações aleatórias
(8)Somar a lista de frações aleatórias criada
(9)Encontrar a maior fração da lista criada
(0)Sair""")

def verFracao(f):
    return (str(f[0]) + "/" + str(f[1]) )

def mdc(a,b):
    if a < b :
        return mdc(b,a)
    elif a % b == 0 :
        return b    
    else:
        return mdc(a, a%b)
        
def criarfracao():
    num = int(input("Qual o numerador? "))
    den = int(input("Qual o denominador? "))
    f = (num, den)
    return f 

def simplificarFracao(f):
    num, den = f
    MDC = mdc (num, den)
    f = (int(num/MDC), int(den/MDC))
    return f

def somarDuas():
    desejo = input("""Quer:
(1)Somar uma fração com a criada na opção 1, caso o tenha feito?
(2)Somar duas frações novas?
""")
    if desejo == "1":
        f2 = criarfracao()
        fr = (f[0]*f2[1] + f[1]*f2[0], f[1]*f2[1])
        r = simplificarFracao(fr) 
        return r
    else:
        f1 = criarfracao() 
        f2 = criarfracao()
        fr = (f1[0]*f2[1] + f1[1]*f2[0], f1[1]*f2[1])
        r = simplificarFracao(fr) 
        return r
        
def subtrairDuas():
    desejo = input("""Quer:
(1)Subtrair uma fração à criada na opção 1, caso o tenha feito?
(2)Subtrair duas frações novas?
""")
    if desejo == "1":
        f2 = criarfracao()
        fr = (f[0]*f2[1] - f[1]*f2[0], f[1]*f2[1])
        r =simplificarFracao(fr) 
        return r
    else:
        f1 = criarfracao() 
        f2 = criarfracao()
        fr = (f1[0]*f2[1] - f1[1]*f2[0], f1[1]*f2[1])
        r = simplificarFracao(fr) 
        return r
    
def multiplicarDuas():
    desejo = input("""Quer:
(1)Multiplicar uma fração à criada na opção 1, caso o tenha feito?
(2)Multiplicar duas frações novas?
""")
    if desejo == "1":
        f2 = criarfracao()
        fr = (f[0]*f2[0], f[1]*f2[1])
        r =simplificarFracao(fr) 
        return r
    else:
        f1 = criarfracao() 
        f2 = criarfracao()
        fr = (f1[0]*f2[0], f1[1]*f2[1])
        r = simplificarFracao(fr) 
        return r 
        
def dividirDuas():
    desejo = input("""Quer:
(1)Dividir uma fração à criada na opção 1, caso o tenha feito?
(2)Dividir duas frações novas?
""")
    if desejo == "1":
        f2 = criarfracao()
        fr = (f[0]*f2[1], f[1]*f2[0])
        r =simplificarFracao(fr) 
        return r
    else:
        f1 = criarfracao() 
        f2 = criarfracao()
        fr = (f1[0]*f2[1], f1[1]*f2[0])
        r = simplificarFracao(fr) 
        return r 

import random
def LfA():
    L = []
    n = int(input("Quantas frações quer na lista? "))
    i = 0
    while i in range (n):
        fa = (random.randint(1,100), random.randint(1,100))
        L.append(simplificarFracao(fa))
        i = i+1
    return L  
    
def sL(L):
    res = (0,1)
    for elem in L:
        res = (res[0]*elem[1] + res[1]*elem[0], res[1]*elem[1])
    res = simplificarFracao(res) 
    return res
    
def MfL(L):
    maximo = L[0]
    for fracao in L:
        N1 = int (maximo[0]*fracao[1])
        N2 = int (fracao[0]*maximo[1])
        if N1 > N2:
            maximo = maximo
        else:
            maximo = fracao
    return maximo      
    
```
